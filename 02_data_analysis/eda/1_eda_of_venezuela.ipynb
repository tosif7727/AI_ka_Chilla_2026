```python
# Threshold: drop columns with >50% missing
threshold = 0.5
missing_ratio = data.isnull().mean()  # fraction of missing per column
cols_to_drop = missing_ratio[missing_ratio > threshold].index
data = data.drop(cols_to_drop, axis=1)

print(f"âœ… Dropped columns: {list(cols_to_drop)}")
```

## âš¡ Dropping Columns with High Missing Values

Columns that had **more than 50% missing data** were automatically removed.  
This helps improve data quality by keeping only the columns with sufficient information for analysis and modeling. âœ…


```python
# Columns with missing values
columns_linear_interpolation = [
    'gdp_growth', 
    'unemployment_rate',
    'government_consumption_percent',
    'current_account_balance_percent',
    'fdi_inflows_percent',
    'oil_rent_percent',
    'natural_resource_rent_percent'
]

# Apply linear interpolation
data[columns_linear_interpolation] = data[columns_linear_interpolation].interpolate(method='linear')

# Forward fill first/last if missing at boundaries
data[columns_linear_interpolation] = data[columns_linear_interpolation].fillna(method='ffill').fillna(method='bfill')

# Check results
print("âœ… Missing values after professional filling:")
print(data[columns_linear_interpolation].isnull().sum())
```

## **ðŸ” Insights & Analysis**

### 1ï¸âƒ£ Oil Dependency Drives the Economy
Venezuelaâ€™s economic performance closely follows **oil and natural resource rents**. Periods of high oil revenue coincide with stronger GDP growth, improved current account balances, and higher exports. This confirms a **mono-commodity economic structure**.

### 2ï¸âƒ£ Investment-Led but Fragile Growth
GDP growth shows a positive relationship with **gross capital formation and imports**, indicating that economic expansion depends on investment and imported inputs. However, this growth is **not structurally sustainable** due to weak domestic production.

### 3ï¸âƒ£ Weak Employment Transmission
Despite periods of GDP recovery, **unemployment remains weakly correlated with growth**, suggesting limited job creation and structural labor market inefficiencies.

### 4ï¸âƒ£ External Vulnerability
Trade indicators and current account balances are highly volatile, making the economy **exposed to external shocks**, especially oil price fluctuations and global demand changes.

### 5ï¸âƒ£ Low Foreign Investor Confidence
FDI inflows remain low even during recovery periods, reflecting **policy uncertainty, institutional risk, and macroeconomic instability**.

### 6ï¸âƒ£ Crisis Impact (2015â€“2020)

- Sharp GDP collapse

Recovery after 2021 remains **partial and fragile**.
