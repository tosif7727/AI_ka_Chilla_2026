```python
# import libraries for anaysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# load dataset
df = pd.read_csv("../../06_datasets/remittances_dataset.csv")

# 2. Inspect the first rows
df.head()
```

```python
# 3. Filter only the **Total Cash Inflow** row
# Based on 'Series Display Name' or 'Series name'
df_total = df[df['Series Display Name'] == 'I. Cash Flow']
```

```python
# 4. Keep only necessary columns
df_total = df_total[['Observation Date', 'Observation Value', 'Unit']]
```

```python
# 5. Rename columns for clarity
df_total = df_total.rename(columns={
    'Observation Date': 'Date',
    'Observation Value': 'Remittances_Million_USD',
    'Unit': 'Unit'
})
```

```python
# 6. Convert 'Date' to datetime
df_total['Date'] = pd.to_datetime(df_total['Date'])
```

```python
# 7. Sort by Date
df_total = df_total.sort_values('Date').reset_index(drop=True)
```

```python
# 8. Optional: convert Million USD to USD
df_total['Remittances_USD'] = df_total['Remittances_Million_USD'] * 1_000_000
```

```python

# 9. Preview cleaned dataset
print(df_total.head())
print(df_total.tail())
```
